```python
importlib.reload(active_learning)
importlib.reload(mice)



test_m_all_iter = []
for m in range(2,7):
    test_m_unseen = []
    for i in range(10):

        X = all_X[i]
        train_idx = train_indices[i]
        test_idx = test_indices[i]

        print(f'iter= {i}, m= {m}')

        test_m_unseen_acc = active_learning.ActiveLearningWithImputationWithUncertainty(X, y, train_idx, test_idx, \
                                                                                  imputation_method='mice', imputation_m=m, \
                                                                                  seed = 174, batch_k=2 ,\
                                                                                  query_method = 'cumir', end_portion=0.1
                                                                                  )
        test_m_unseen.append(test_m_unseen_acc)

    test_m_all_iter.append(test_m_unseen)
```

## Graph

```python
import matplotlib.pyplot as plt
```

```python
# x = np.arange()
c = ['cornflowerblue', 'mediumaquamarine', 'mediumorchid', 'hotpink', 'forestgreen']

for i in range(2,7):
    idx = i - 2

    legend = f'm= {i}'

    ys = np.array(test_m_all_iter[idx])
    ys_mean = ys.mean(axis=0)
    round_std = np.std(ys, axis=0)

    plt.plot(ys_mean, color=c[idx], label = legend)
    x = np.arange(len(ys_mean))
    plt.errorbar(x, ys_mean, yerr=round_std, fmt='none', color=c[idx], alpha=0.1, capsize=1)

    plt.legend()

plt.xlabel('Simulation')
plt.ylabel('Model accuracy')
plt.title('Model accuracy for MICE and CUMIR \nwith different number of imputation simulation')
plt.savefig(f'{root}/figs/optimize_m.png')

plt.show()
```
